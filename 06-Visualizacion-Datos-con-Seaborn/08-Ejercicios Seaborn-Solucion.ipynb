```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.distplot(titanic['fare'],kde=False,bins=30, color="red")
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.boxplot(x="class", y="age", data=titanic,palette='rainbow')
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.swarmplot(x="class", y="age", data=titanic, palette="deep")
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.countplot(x='sex',data=titanic)
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
sns.heatmap(titanic.corr(),cmap='coolwarm')
```

```python
# CÓDIGO AQUÍ
# REPLICAR EL EJERCICIO DE LA IMAGEN DEL GRUPO A CONTINUACIÓN
# TENER CUIDADO DE NO SOBRESCRIBIR LA CELDA ABAJO
# ¡QUE ELIMINARÍA LA IMAGEN DEL EJERCICIO!
g = sns.FacetGrid(titanic, col="sex")
g = g.map(plt.hist, "age")
```
