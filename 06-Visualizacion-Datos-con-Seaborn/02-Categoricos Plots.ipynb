# Gráficos de datos categóricos

¡Ahora analicemos el uso de seaborn para trazar datos categóricos! Hay algunos tipos de graficos principales para esto:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

¡Repasemos ejemplos de cada uno!

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## barplot y countplot

Estos gráficos muy similares permiten obtener datos agregados de una característica categórica en sus datos. **barplot** es un gráfico general que permite agregar los datos categóricos basados en alguna función, por defecto la media:

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

Puede cambiar el objeto estimador a su propia función, que convierte un vector en un escalar:

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### countplot

Esto es esencialmente lo mismo que el diagrama de barras, excepto que el estimador está contando explícitamente el número de ocurrencias. Es por eso que solo pasamos el valor x:

```python
sns.countplot(x='sex',data=tips)
```

## boxplot and violinplot

Se utilizan diagramas de caja y diagramas de violín para mostrar la distribución de datos categóricos. Un diagrama de caja (o diagrama de caja y bigotes) muestra la distribución de datos cuantitativos de una manera que facilita las comparaciones entre variables o entre niveles de una variable categórica. El cuadro muestra los cuartiles del conjunto de datos mientras que los bigotes se extienden para mostrar el resto de la distribución, excepto para los puntos que se determinan como "valores atípicos" mediante un método que es una función del rango intercuartil.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# Se puede hacer todo el dataframe con orient = 'h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

### violinplot

Una grafica de violín juega un papel similar a una trama de caja y bigotes. Muestra la distribución de datos cuantitativos en varios niveles de una (o más) variables categóricas de modo que esas distribuciones se puedan comparar. A diferencia de un diagrama de caja, en el que todos los componentes del diagrama corresponden a puntos de datos reales, el diagrama de violín presenta una estimación (probabilidad) de la densidad.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

## stripplot y swarmplot

La gráfica de tira dibujará una gráfica de dispersión donde una variable es categórica. Se puede dibujar un diagrama de bandas por sí solo, pero también es un buen complemento para un diagrama de caja o violín en los casos en los que desee mostrar todas las observaciones junto con alguna representación de la distribución subyacente.

El **swarmplot()** es similar a **stripplot()**, pero los puntos se ajustan (solo a lo largo del eje categórico) para que no se superpongan. Esto proporciona una mejor representación de la distribución de valores, aunque no se escala tan bien a un gran número de observaciones (tanto en términos de la capacidad de mostrar todos los puntos como en términos del cálculo necesario para organizarlos).

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split = True)
```

### Combinando gráficos categóricos

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

## factorplot

**factorplot** es la forma más general de un diagrama categórico. Puede tomar un parámetro **kind** para ajustar el tipo de gráfico:

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```
